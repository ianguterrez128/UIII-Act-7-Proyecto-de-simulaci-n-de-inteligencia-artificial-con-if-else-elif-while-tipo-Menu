```python
# ===================================
# FARMACEUTICO
# ===================================

print("===================================")
print(" FARMACEUTICO ")
print("===================================")

nombre = input("Ingrese el nombre del paciente: ")
edad = int(input("Ingrese la edad: "))

while True:

    print("\n=========== MENU ===========")
    print("1. Revisar sintomas")
    print("2. Revisar medicamentos")
    print("3. Analizar paciente")
    print("4. Salir")

    opcion = int(input("Seleccione una opcion: "))

    # -----------------------------------
    # OPCION 1
    # -----------------------------------

    if opcion == 1:

        temperatura = float(input("Ingrese la temperatura corporal: "))
        tos = input("¿Tiene tos? (si/no): ")
        dolor = input("¿Tiene dolor de cabeza? (si/no): ")

        print("\nANALIZANDO SINTOMAS...\n")

        if temperatura >= 39 and tos == "si":
            print("Posible infeccion fuerte.")
            print("Medicamento recomendado: Paracetamol")

        elif temperatura >= 38:
            print("Posible gripe.")
            print("Medicamento recomendado: Antigripal")

        elif dolor == "si":
            print("Posible migraña o cansancio.")
            print("Medicamento recomendado: Ibuprofeno")

        else:
            print("No se detectan sintomas graves.")
            print("Paciente estable.")

    # -----------------------------------
    # OPCION 2
    # -----------------------------------

    elif opcion == 2:

        alergia = input("¿Es alergico a la penicilina? (si/no): ")
        presion = input("¿Tiene presion alta? (si/no): ")

        print("\nREVISANDO MEDICAMENTOS...\n")

        if alergia == "si":
            print("No administrar penicilina.")

        elif presion == "si":
            print("Evitar medicamentos fuertes.")

        else:
            print("Medicamentos normales permitidos.")

    # -----------------------------------
    # OPCION 3
    # -----------------------------------

    elif opcion == 3:

        print("\nANALISIS DEL PACIENTE...\n")

        if edad >= 60:
            print("Paciente de la tercera edad.")
            print("Se recomienda revision constante.")

        elif edad < 12:
            print("Paciente menor de edad.")
            print("Usar dosis infantiles.")

        else:
            print("Paciente estable.")

    # -----------------------------------
    # OPCION 4
    # -----------------------------------

    elif opcion == 4:

        print("\n===================================")
        print(" REPORTE FINAL ")
        print("===================================")

        print("Paciente:", nombre)
        print("Edad:", edad)
        print("Programa realizado por Ian Gutierrez NL=26")

        print("\nGracias por utilizar FARMACEUTICO IA")

        break

    # -----------------------------------
    # OPCION INVALIDA
    # -----------------------------------

    else:

        print("Opcion no valida")
```
